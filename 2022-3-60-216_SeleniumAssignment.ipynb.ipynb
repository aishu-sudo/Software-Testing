## TASK 3 — Scrape Pages 1–4

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Start fresh browser for scraping
driver = webdriver.Chrome()
driver.get("https://books.toscrape.com")
time.sleep(2)

books_data = []

for page in range(1, 5):
    print(f"\n📄 Scraping Page {page}")

    books = driver.find_elements(By.CLASS_NAME, "product_pod")

    for book in books:
        # Title (use 'title' attribute for full, non-truncated title)
        title = book.find_element(By.TAG_NAME, "h3").find_element(By.TAG_NAME, "a").get_attribute("title")

        # Price
        price = book.find_element(By.CLASS_NAME, "price_color").text

        # Availability
        availability = book.find_element(By.CLASS_NAME, "availability").text.strip()

        # Rating (extract second word)
        rating_class = book.find_element(By.CLASS_NAME, "star-rating").get_attribute("class")
        rating = rating_class.split()[1]

        books_data.append({
            "page": page,
            "title": title,
            "price": price,
            "rating": rating,
            "availability": availability
        })

    # Go to next page
    if page < 4:
        driver.find_element(By.XPATH, "//li[@class='next']/a").click()
        time.sleep(2)
```

```python
import csv

print("\nTotal books collected:", len(books_data))

# Assertion
if len(books_data) == 80:
    print("PASS: Correct number of books (80)")
else:
    print(" FAIL: Expected 80 books")

# Save CSV
with open("books_data.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["page", "title", "price", "rating", "availability"])
    writer.writeheader()
    writer.writerows(books_data)

print("CSV saved as books_data.csv")

# Show first 3 rows
print("\nFirst 3 rows:")
for row in books_data[:3]:
    print(row)
```

```python
total_tests = 0
passed_tests = 0

def log(test_name, passed):
    global total_tests, passed_tests
    total_tests += 1

    if passed:
        passed_tests += 1
        print(f" PASS: {test_name}")
    else:
        print(f" FAIL: {test_name}")
```

```python
# Re-initialize browser (in case previous session expired)
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
driver.get("https://books.toscrape.com")
time.sleep(2)

# Pick first 3 books
sample_books = books_data[:3]

for book in sample_books:
    title = book["title"]

    print(f"\n Testing book: {title}")

    # Find and click book
    link = driver.find_element(By.PARTIAL_LINK_TEXT, title[:10])
    link.click()
    time.sleep(2)

    # a. Page title not empty
    log("Page title not empty", driver.title != "")

    # b. Title matches
    page_title = driver.find_element(By.TAG_NAME, "h1").text
    log("Title matches", title[:10] in page_title)

    # c. Price matches
    page_price = driver.find_element(By.CLASS_NAME, "price_color").text
    log("Price matches", page_price == book["price"])

    # d. Add to basket button
    try:
        btn = driver.find_element(By.CSS_SELECTOR, "button.btn-primary[type='submit']")
        log("Add to basket enabled", btn.is_enabled())
    except:
        log("Add to basket exists", False)

    # e. Availability
    availability = driver.find_element(By.CLASS_NAME, "availability").text
    log("In stock check", "In stock" in availability)

    # Go back
    driver.back()
    time.sleep(2)
```
